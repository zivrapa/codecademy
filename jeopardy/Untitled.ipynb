```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.read_csv('jeopardy.csv')
df.rename(columns = {'Show Number': 'Show_Number', ' Air Date': 'Air_Date', ' Category': 'Category', ' Value': 'Value', ' Question': 'Question', ' Answer': 'Answer',' Round': 'Round'}, inplace = True) 
print(df.head())
df.columns
```

```python
words_list = ['king', 'England']

def words_list_occ(data, list_of_words):
    filter = lambda question: all(word.lower() in question.lower() for word in list_of_words)
    return data.loc[data["Question"].apply(filter)]
words_list_occ(df, words_list)
```

```python
df['Value'] = df.Value.apply(lambda x: x.replace('$','').replace(',','') if x != 'None' else 0)
df['float_value'] = df['Value'].astype(float)

```

```python
df['float_value'].mean()
```

```python
filter_data = words_list_occ(df, 'king')
mean_value_filter_data = filter_data['float_value'].mean() 
mean_value_filter_data
```

```python
double = df[df['Round'] == 'Double Jeopardy!']
double.head()
double['float_value'].mean()
```

```python
not_double = df[df['Round'] == 'Jeopardy!']
not_double['float_value'].mean()
```
